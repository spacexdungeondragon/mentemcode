```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("size:", arr.size, "data type:", arr.dtype, "dimensions:", arr.ndim)
print("reshaped array:", np.reshape(arr, (5, 1)))

arr2 = np.array([6, 7, 8, 9, 10])

print("vertically stack arr and arr2:", np.vstack((arr,arr2)))
```
