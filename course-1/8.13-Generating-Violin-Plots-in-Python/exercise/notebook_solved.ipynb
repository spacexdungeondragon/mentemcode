```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_excel('Customer-Churn-Records.xlsx')
```

## Task 1

```python
plt.figure(figsize=(10, 6))
 
sns.violinplot(y='Age', data=df, orient='v', width=0.7)
 
plt.title("Violin plot of Age")
plt.xlabel("Violin plot")
plt.ylabel("Age")
 
plt.show()
```

## Task 2


```python
plt.figure(figsize=(10, 6))
 
sns.violinplot(x='Gender', y='Age', data=df, orient='v', width=0.7, palette='Pastel1')
 
plt.title("Violin plot of Age by Gender")
plt.xlabel("Violin plot")
plt.ylabel("Age")
 
plt.show()
```

## Task 3

```python
plt.figure(figsize=(10, 6))
 
sns.violinplot(x='Gender', y='Age', hue='Complain', split=True, 
               data=df, orient='v', width=0.7, palette='Pastel1')
 
plt.title("Violin plot of Age by Gender")
plt.xlabel("Violin plot")
plt.ylabel("Age")
 
plt.show()
```

## Task 4


```python
plt.figure(figsize=(10, 6))
 
sns.violinplot(x='Gender', y='Age', hue='Complain', split=False, 
               data=df, orient='v', width=0.7, palette='Pastel1')
 
plt.title("Violin plot of Age by Gender")
plt.xlabel("Violin plot")
plt.ylabel("Age")
 
plt.show()
```
