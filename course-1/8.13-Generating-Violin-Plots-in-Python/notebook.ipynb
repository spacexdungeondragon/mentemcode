To help you generate your first violin plot, `Matplotlib.plyplot` has been imported as plt and Seaborn has been imported as sns. The pre-loaded data frame `df` contains the data from **Customer-Churn-Records**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_excel('Customer-Churn-Records.xlsx')
```

The following code generates a violin plot using `Matplotlib` and `Seaborn` in combination. Here's what each part of the `violinplot()` statement does:


```python
plt.figure(figsize=(10, 6)) 
sns.violinplot(y='Balance', data=df, orient='v', width=0.7)
 
plt.title("Violin plot of Balance")
plt.xlabel("Violin plot")
plt.ylabel("Balance")
 
plt.show()
```

- `sns.violinplot()` calls the violinplot function from the Seaborn library (which is commonly imported as sns). This function creates a violin plot, which is a combination of a box plot and a kernel density plot.
 
- `y='Balance'` specifies that the variable to be plotted on the y-axis is `'Balance'`. This variable comes from the data frame `df`.
 
- `data=df` specifies the data source, which is a Pandas data frame named `df`.
 
- `orient='v'` sets the orientation of the plot to vertical. If you wanted a horizontal plot, you could use `'h'` instead. When you change between vertical and horizontal orientation, your `x` and `y` variables have to change accordingly.
 
- `width=0.7` sets the width of the violin plot. The width determines how wide the `‘body’` of the violin plot will be.
 
- `palette='Set1'` sets the colour palette of the plot to `'Set1'`, which is a predefined set of colours in `Seaborn`.

## Multiple violin plots

You can create multiple violin plots side-by-side to compare different categories.
 
To do this, simply add a categorical `x` variable to the `violinplot()` statement. The following code splits the y-variable `'Balance'` across the three geographies in your dataset. Run the code to see the output:

```python
plt.figure(figsize=(10, 6))
 
sns.violinplot(x='Geography', y='Balance', data=df, 
               orient='v', width=0.7, palette='Set1', hue='Geography', legend=False)
 
plt.title("Split Violin plot of Balance by Geography")
plt.xlabel("Geography")
plt.ylabel("Balance")
 
plt.show()
```

Violin plots allow you to further enhance the visualisation of your data by adding a third dimension. This can be particularly helpful when you want to compare the distribution of two sub-populations across the data. 
 
Let's use this feature to explore whether the distribution of account balances is different between genders.
 
In the following code, the parameter `hue='Gender'` determines gender as the third dimension, while `split=True` creates a violin plot that shows the gender distribution as the left and right halves of the body of the plot: 

```python
plt.figure(figsize=(10, 6))
 
sns.violinplot(x='Geography', y='Balance', data=df, hue='Gender',
               split=True, orient='v', width=0.7, palette='Set1')
 
plt.title("Split Violin plot of Balance by Geography")
plt.xlabel("Geography")
plt.ylabel("Balance")
 
plt.show()
```

As you can see, while the density of balances is quite different between states, it is very similar between genders.
