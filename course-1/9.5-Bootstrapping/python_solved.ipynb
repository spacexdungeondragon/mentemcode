```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([85, 90, 78, 92, 88, 76, 84, 89, 82, 87])
# Number of resamples
n_resamples = 1000

# Create an empty array to store the means of resamples
resample_means = []

# Perform bootstrapping
for _ in range(n_resamples):
    resample = np.random.choice(data, size=len(data), replace=True)
    resample_mean = np.mean(resample)
    resample_means.append(resample_mean)
    
# Plot the distribution of resample means
plt.hist(resample_means, bins=30, edgecolor='k')
plt.title('Bootstrapped Sample Means')
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
plt.show()
```
