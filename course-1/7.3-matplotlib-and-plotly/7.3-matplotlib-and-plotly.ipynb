Matplotlib Exercise: Average Monthly Temperatures in Melbourne (Celsius)
For this exercise, you'll be working with a dataset containing the average monthly temperatures in Melbourne. Your task is to create a line chart to visualise these temperatures in the "main.py" notebook.

Your tasks:

Import the necessary libraries 

Create a line plot with month on the x-axis and temperature on the y-axis

Add a title to your plot - "Average Monthly Temperatures in Melbourne"

Add labels to the x and y axes - "Months" and "Temperature" respectively 

Display your plot using  plt.savefig('plot.png').

If you get stuck, refer to "main_solved.py" to help with coding in "main.py."

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May",
                        "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperatures_in_celsius = [26, 26, 24, 20, 16, 14, 14, 15, 17, 20, 22, 24]

plt.plot(months, temperatures_in_celsius)
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Average Monthly Temperatures in Melbourne')

plt.savefig('plot.png')
```

Plotly Exercise: Global Sales of Video Game Genres (in millions)
In this exercise, you'll be working with a dataset containing the global sales of different video game genres. Your task is to create an interactive bar chart to visualise these sales in the "main.py" notebook.

Your tasks:

Import the necessary libraries

Create a bar chart with 'genres' on the x-axis and 'sales' on the y-axis

Add a title to your plot - "Global Sales of Video Game Genres"

Add labels to the x and y axes - "Genres" and "Sales (in millions)" respectively

Customise the layout and styling of your plot as you like

Display your plot using fig.write_image('plot.png').

If you get stuck, refer to "main_solved.py" to help with coding in "main.py."

```python
import plotly.express as px
import pandas as pd

genres = ["Action", "Sports", "Shooter", "Role-Playing", "Platform",
                     "Racing", "Fighting", "Strategy", "Adventure", "Puzzle"] 
sales = [1751, 1330, 1052, 934, 831, 732, 448, 174, 239, 244]

df = pd.DataFrame({
    "Genre": genres,
    "Sales (in millions)": sales,
})

fig = px.bar(df, x="Genre", y="Sales (in millions)", title="Global Sales of Video Game Genres")
fig.write_image('plot.png')
```
