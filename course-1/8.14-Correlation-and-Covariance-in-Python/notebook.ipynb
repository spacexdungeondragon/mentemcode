```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

stocks = pd.read_excel('stock_index.xlsx')
```

For this topic, you will be using the data frame `stocks`, which contains daily returns of various stock indices from 5 January 2009 to 22 February 2011. 
 
This data frame has already been loaded for you.
 
In Python, the `cov()` function is used to calculate covariance. Here's how you can use it:


```python
 print(stocks.cov())
```

This will give you the covariance matrix, which shows the covariance between all pairs of numeric variables – in this case stock indices – in your dataset. 
 
The generated covariance matrix is not particularly useful for anything else than concluding that all indices in the dataset have a positive covariance.
 
Therefore, let's turn to a correlation matrix for a more useful view.
 
The `corr()` function is used to calculate correlation. Here's how you can use it:

```python
print(stocks.corr())
```

This will give you the correlation matrix, which shows the correlation between all pairs of indices in `stocks`. This is a more useful view than covariance, as the correlation shows the relative strength of each variable pair.
 
Note that Python will generate covariance and correlation matrices between all numeric variables, even if these combinations are not logically correct. 
 
As you begin to use and interpret correlation matrices, be careful not to misinterpret correlations between non-continuous variables that are numeric in nature, such as binary flags `(1, 0)`, customer numbers, phone numbers and ordinal numbers with a small number of categorical outcomes.
 
In short, always think about whether correlations make logical sense before drawing inferences from them.
 
Although the basic correlation matrix generated by the Pandas function `corr()` is very useful, it is not visually pleasing or easy to read. Therefore, it is beneficial to explore two methods for visualising correlations: scatter plots and correlation matrix heatmaps.
 
Firstly, let's use a scatter plot to visualise the relationship between two highly correlated indices in the `stocks` data frame, the British FTSE index and the German DAX index. This pair has a correlation of `0.87`.
 
Generating scatter plots is a very simple exercise using Matplotlib.pyplot's `scatter()` command:
 

```python
plt.scatter(stocks['FTSE'], stocks['DAX'])
plt.title('Relationship between FTSE and DAX')
plt.xlabel('FTSE')
plt.ylabel('DAX')
plt.show()
```

Let's now generate a correlation matrix and overlay it with a heatmap. For this exercise you will use the `heatmap()` chart type from Seaborn. 
 
Here's what's happening in the code: 
- The Seaborn function `heatmap()` takes the generated correlation matrix, `corr_matrix`, and turns it into a heatmap.
- The argument `annot=True` means that the values of the correlations will be written on each cell of the heatmap.


```python
corr_matrix = stocks.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

As you can see from the output, heatmaps make it very easy to spot the strength of relationships across a multitude of variables. 
 
This sort of view is a key tool for data scientists, as it allows you to identify important relationships across complex datasets. Data scientists will use correlation heat maps to spot key relationships to test in predictive models, but also to identify any multicollinearity that might cause issues in a model.
