We will use `Matplotlib.pyplot` (as plt) to create histograms
and the Seaborn library (as `sns`) to create density plots. **Numpy** and **Pandas** have also been imported for you.

For this example, you will be using the **Customer-Churn-Records** dataset, which has been preloaded as the data frame `df`.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read customer churns records

df = pd.read_excel('cust_churn.xlsx')
```

## Generate a histogram of credit scores

First, create a histogram of the variable *CreditScore*. You want to create a relative frequency histogram so that the y-axis is comparable with the density plot.

Let’s break down what’s happening in the Python code:

1. `plt.figure(figsize=(8, 6))`: This line is setting up the figure where the histogram will be plotted. The argument figsize=(8, 6) sets the size of the figure to be 8 inches wide and 6 inches tall.
 
2. `plt.hist(df['CreditScore'], bins=20, color="blue", density=True, alpha=0.7)`: This line is doing the actual plotting of the histogram. It's taking data from the 'CreditScore' column of `df` and organising it into 20 bins (specified by bins=20). It's colouring the bars of the histogram blue (color="blue"), with a transparency level of 70% (alpha=0.7). 
 
3. The `density=True` part means that the histogram is normalised, so the sum of all the bar heights equals 1 - which allows it to represent a probability density.
 
4. `plt.title("Histogram")`: This line is giving the plot a title of ‘Histogram’.
 
5. `plt.xlabel("Value")` and `plt.ylabel("Relative frequency")`: These lines are labelling the x-axis and y-axis as `'Value'` and `'Relative frequency'`, respectively.
 
6. `plt.show()`: Finally, this line is displaying the plot. Until this command is called, all the commands you've been giving (like setting the title and labels) have just been recorded by `matplotlib`,but not actually shown on the screen.

As you can see, the distribution of credit scores is skewed towards higher values.


```python
# Create a histogram and density plot
plt.figure(figsize=(8, 6))
plt.hist(df['CreditScore'], bins=20, color="blue", density=True, alpha=0.7)
sns.kdeplot(df['CreditScore'], color="black", fill=False)
plt.title("Histogram and density")
plt.xlabel("Value")
plt.ylabel("Relative frequency/density")
plt.show()
```

As you can see, the PDF is working in the background to smooth out any peaks and valleys by estimating an underlying distribution.
 
## Customising Density Plots
 
**Seaborn** allows you to customise your density plots. For example, you can change the bandwidth of the KDE to make the plot smoother or rougher.
 
By default, the bandwidth is equal to `1`, so anything larger than `1` will smoothen the curve and anything smaller than `1` will make it rougher. 
 
In the following code you create an additional two density plots with bandwidth adjustments of `0.2 `(the red line) and `3` (the orange line) respectively. You can think of these adjustment factors as relative coefficients that arbitrarily adjust the bandwidth.
 
In this case, both adjustments are quite large. The reduced bandwidth introduces a lot of additional noise into the distribution, whereas the increased bandwidth smooths out the distribution too much, thereby removing potential insights. Both adjustments also introduce the perception that credit scores can take values beyond `850`, which is untrue. 

```python
 # Create a histogram and 3 density plots
plt.figure(figsize=(8, 6))
plt.hist(df['CreditScore'], bins=20, color="blue", density=True, alpha=0.7)
sns.kdeplot(df['CreditScore'], color="black", fill=False)
sns.kdeplot(df['CreditScore'], color="red", bw_adjust=0.2, fill=False)
sns.kdeplot(df['CreditScore'], color="orange", bw_adjust=3, fill=False)
plt.title("Histogram and density")
plt.xlabel("Value")
plt.ylabel("Relative frequency/density")
plt.show()
```

A common mistake when creating density plots is choosing an inappropriate bandwidth. A bandwidth that is too small may create a plot with too much noise, while a bandwidth that is too large may oversmooth the plot, obscuring important details.
 
Another mistake is not considering the underlying distribution of your data. While *KDE* does not assume any specific distribution, different distributions may require different considerations in terms of bandwidth and kernel choice.
 
These adjustments are often chosen on the basis of heuristics, so determining the right adjustment (if any) to bandwidth and distribution is somewhat trial and error.
