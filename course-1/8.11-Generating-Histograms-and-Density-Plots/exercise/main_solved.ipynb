```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read customer churns records
df = pd.read_excel('cust_churn.xlsx')

# Set up the figure
plt.figure(figsize=(8, 6))

# Create a histogram of the 'Age' column
plt.hist(df['Age'], bins=30, color='green', density=True, alpha=0.7)

# Add labels and title
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Relative frequency/density')
 
# Create a density plot of the 'Age' column
sns.kdeplot(df['Age'], color='black', bw_adjust=2)

# Show the plot
plt.show()
```
