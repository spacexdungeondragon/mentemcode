```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read customer churns records
df = pd.read_excel('cust_churn.xlsx')
```

Let's get started with generating a box plot using Matplotlib. You will start by plotting the `'Balance'` variable from the Customer-Churn-Records dataset, which has been preloaded as the data frame `df`.
 
Matplotlib has also been imported for you.
 
To refresh your memory, you previously generated a histogram of the `'Balance'` column, which looked like this:
![histogram](histogram.png)

Now, letâ€™s generate a box plot for the same data:

```python
plt.figure(figsize=(10, 6)) 
plt.boxplot(df['Balance'], vert=True, widths=0.7, patch_artist=True,
            boxprops=dict(facecolor="blue", alpha=0.7))
 
plt.title("Box Plot of Balance")
plt.xlabel("Balance")
 
plt.show()
```

Here's a step-by-step breakdown of the code:

1. `plt.figure(figsize=(10, 6))`: This line is used to create a new figure and set the size of the figure to be `10` units wide and `6` units tall. It's optional and if not specified, Matplotlib uses default values.
 
2. `plt.boxplot(df['Balance'], vert=True, widths=0.7, patch_artist=True, boxprops=dict(facecolor="blue", alpha=0.7))`: This line generates the box plot with several customisation parameters:
`df['Balance']` specifies that the data for the box plot comes from the 'Balance' column in the data frame `df`.
`vert=True` means that the box plot will be vertical. If it were False, the box plot would be horizontal.

3. `widths=0.7` sets the width of the box in the box plot.
4. `patch_artist=True` allows the box to be filled with colour.

5. `boxprops=dict(facecolor="blue", alpha=0.7)` is a dictionary specifying properties for the box in the box plot. In this case, the box's face colour is set to blue with an alpha (transparency) value of `0.7`.
 
6. `plt.title("Box Plot of Balance")`: This line adds the title `'Box Plot of Balance'` to the plot.
 
7. `plt.xlabel("Balance")`: This line labels the x-axis as â€˜Balanceâ€™.
 
8. `plt.show()`: This line displays the plot. Until this function is called, the plot is actually not rendered.
 

## Creating multiple box plots using Seaborn

Box plots can be especially useful when they are displayed side by side to compare the distributions of a variable across other categories. For this particular purpose, the Seaborn syntax is a lot more user-friendly than Matplotlibâ€™s.
 
Let's generate box plots that display account balances by geography, using Seaborn.
 
To do this, you use the variable `'Geography'`, which contains the three states NSW, QLD and VIC, as the class variable. `'Balance'` contains the data being displayed in the box plots.
 
Now, letâ€™s generate a box plot for the same data:

```python
plt.figure(figsize=(10, 6))  
sns.boxplot(x='Geography', y='Balance', data=df, width=0.7, hue='Geography')
 
plt.title("Box Plots of Balance by Geography")
plt.xlabel("Geography")
plt.ylabel("Balance")
 
plt.show()
```

By visualising three box plots side by side, you can easily compare the distribution of account balances by state. You can now see that all the zero-balance accounts are in NSW and QLD, and that the distribution of account balances is much tighter in VIC.
 
In Seaborn, the function `sns.boxplot()` creates the box plot. The parameters and interpretation remain the same as in Matplotlib.

Here's what each part of `sns.boxplot(x='Geography', y='Balance', data=df, width=0.7, palette='Set3')` does:

1. `x='Geography' and y='Balance'`: These parameters set 'Geography' as the categorical variable (x-axis) and `'Balance'` as the numerical variable (y-axis). In other words, this box plot will show the distribution of `'Balance'` for each category in `'Geography'`.
2. `data=df`: This parameter specifies that the data for the box plot comes from the data frame `df`.
3. `width=0.7`: This parameter sets the width of each box in the box plot.
4. `palette='Set3'`: This parameter sets the colour palette for the boxes in the box plot. `'Set3'` is one of the colour palettes provided by Seaborn.
 
Note: Seaborn has six variations of its default colour palette: deep, muted, pastel, bright, dark, and colorblind.
