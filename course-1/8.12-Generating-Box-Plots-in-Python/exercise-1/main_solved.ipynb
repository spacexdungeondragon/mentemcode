```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read customer churns records
df = pd.read_excel('Customer-Churn-Records.xlsx')
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['CreditScore'], vert=False, widths=0.5, patch_artist=True,
            boxprops=dict(facecolor="green", alpha=0.6))
plt.title("Box Plot of CreditScore")
plt.xlabel("CreditScore")
plt.show()
```
