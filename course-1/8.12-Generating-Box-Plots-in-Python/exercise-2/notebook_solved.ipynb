```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Read customer churns records
df = pd.read_excel('Customer-Churn-Records.xlsx')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Geography', y='CreditScore', data=df, width=0.7, hue='Geography')
plt.title("Box Plots of CreditScore by Geography")
plt.xlabel("Geography")
plt.ylabel("CreditScore")
plt.show()
```
