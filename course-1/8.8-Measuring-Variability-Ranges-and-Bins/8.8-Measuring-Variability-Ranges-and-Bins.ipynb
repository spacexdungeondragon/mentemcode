Calculating Variance Metrics in Python
Let’s calculate variability metrics on the ‘Balance’ column in the Customer-Churn-Records dataset, using Python in the "main.py" notebook.

This table has already been loaded as the data frame df. Pandas has been imported as pd and matplotlib.pyplot as plt.

What kind of distribution can you see?
The account column actually contains two distributions: Accounts with balances > $0 look like they are approximately normally distributed, but there is also a large number of zero-balance accounts which will skew your metrics.

Let’s calculate the mean, median, range, interquartile range, standard deviation, MAD, medAD and percentiles using Pandas functions.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('cust_churn.xlsx')

plt.figure(figsize=(10, 6))

# Create the histogram
plt.hist(df['Balance'], bins=20, color="blue", alpha=0.7)

plt.title("Histogram of Balance")
plt.xlabel("Balance")
plt.ylabel("Frequency")

plt.savefig('plot.png')

mean = df["Balance"].mean()
median = df["Balance"].median()
range = df["Balance"].max() - df["Balance"].min()
iqr = df["Balance"].quantile(0.75) - df["Balance"].quantile(0.25)
std_dev = df["Balance"].std()
mad = (df["Balance"] - df["Balance"].mean()).abs().mean()
medad = (df["Balance"] - median).abs().median()
pct = df["Balance"].quantile([0.25, 0.5, 0.75])

print(
    "Metrics:\n",
    f"Mean: {mean}\n",
    f"Median: {median}\n",
    f"Range: {range}\n",
    f"IQR: {iqr}\n",
    f"SD: {std_dev}\n",
    f"MAD: {mad}\n",
    f"MedAD: {medad}\n",
    f"Percentiles:\n {pct}\n",
)
```
