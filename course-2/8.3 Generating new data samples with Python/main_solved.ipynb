```python
from sklearn.datasets import make_classification
from matplotlib import pyplot as plt
import numpy as np
from collections import Counter
 
# Create new dataset
X, y = make_classification(n_samples=9000, n_features=2, n_redundant=0,
                           n_clusters_per_class=1, weights=[0.98], flip_y=0, random_state=1)
 
for label in set(y):
    row_ix = np.where(y == label)
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
    
plt.legend()
plt.show()
 
from imblearn.over_sampling import SMOTE
 
# Call the SMOTE method
smote_sample = SMOTE(sampling_strategy=0.4, k_neighbors=5, random_state=45)
 
# Transform the original dataset to x/y coordinates
X_1, y_1 = smote_sample.fit_resample(X, y)
 
# Count the new distribution
counter = Counter(y_1)
print(counter)
 
for label in set(y):
    row_ix = np.where(y_1 == label)
    plt.scatter(X_1[row_ix, 0], X_1[row_ix, 1], label=str(label))
    
plt.legend()
plt.show()
```
