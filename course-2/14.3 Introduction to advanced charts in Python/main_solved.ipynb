```python
import plotly.graph_objects as go
import pandas as pd
 
df = pd.read_csv("HistoricalPricesCBA.csv")
 
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
               open=df['Open'],
               high=df['High'],
               low=df['Low'],
               close=df['Close'])])
 
fig.update_layout(
   title='CBA Stock Prices (Candlestick Chart)',
   yaxis_title='CBA Stock',
   xaxis_title='Price'
)
 
fig.show()
```

Alt text: Candlestick chart created with the historical data of the Commonwealth Bank's stock in December 2023.
